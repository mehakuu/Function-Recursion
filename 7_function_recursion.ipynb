```python
def percent(marks):
    p = ((marks[0] + marks[1] + marks[2]+ marks[3])/400 )*100
    return p
```

```python
marks1 = [45, 78, 86, 77]
percentage1 = percent(marks1)    #function calling
percentage1
```

```python
marks2 = [75, 98, 88, 78]
percentage2 = percent(marks2)
print(percentage1, percentage2)
```

# recursion 

```python
#factorial using recursion
def fact_rec1(n):
    if n==0 or n==1:
        return 1
    return n * fact_rec1(n-1)
```

```python
fact_rec1(5)
```
